## 3. Vizualise Results

```python
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import os

# Load the JSON data
with open(os.path.join(PATH_DATA, 'yolo_predictions.json')) as json_file:
    data = json.load(json_file)

# Display images with predicted regions
for image_id, image_data in data['_via_img_metadata'].items():
    # Extract the base filename (remove the appended numbers)
    base_filename = image_id.split('.')[0] + '.jpg'  # Assuming all images are .jpg
    image_path = os.path.join(PATH_DATA, base_filename)
    
    # Debugging: Print the image ID and path
    print(f"Trying to load image: {base_filename}")
    print(f"Constructed image path: {image_path}")

    # Check if the file exists before trying to open it
    if not os.path.exists(image_path):
        print(f"File not found: {image_path}")
        continue  # Skip to the next image if the file does not exist

    # Load the image
    image = Image.open(image_path)

    # Create a figure and axis
    fig, ax = plt.subplots(1, figsize=(12, 8))
    ax.imshow(image)

    # Draw bounding boxes for each region
    for region in image_data['regions']:
        x = region['shape_attributes']['x']
        y = region['shape_attributes']['y']
        width = region['shape_attributes']['width']
        height = region['shape_attributes']['height']
        
        # Create a rectangle patch
        rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

        # Optionally add class name
        class_name = data['_via_attributes']['region']['Classes']['options'][str(region['region_attributes']['Classes'])]
        ax.text(x, y, class_name, color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))

    # Show the plot
    plt.axis('off')
    plt.title(f"Predictions for {base_filename}")
    plt.show()
```

## 4. Break .json file into XML, one per image

1. Load JSON Data: The code reads the JSON file containing YOLO predictions.
Create Output Directory: It creates a directory for the XML files if it doesn't already exist.
3. Create XML Function: A function create_xml is defined to convert the JSON data for each image into XML format.
4. Iterate and Save: The code iterates through each image's metadata, generates the XML content, and saves it to a file named after the image (without the .jpg extension).
Note:

```python
#!pip install lxml
```

```python
import json
import os
from lxml import etree
from datetime import datetime

# Load the JSON data
with open(os.path.join(PATH_DATA, 'yolo_predictions.json')) as json_file:
    data = json.load(json_file)

# Extract the class mapping from the JSON
class_mapping = {str(key): value for key, value in data['_via_attributes']['region']['Classes']['options'].items()}

# Function to create XML content
def create_xml(image_id, image_data):
    # Create the root element with the specified namespaces
    root = etree.Element(
        "{http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15}PcGts",
        nsmap={
            None: "http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15",
            "xsi": "http://www.w3.org/2001/XMLSchema-instance"
        }
    )
    
    # Add XML declaration
    xml_declaration = '<?xml version="1.0" encoding="UTF-8" standalone="yes"?>\n'
    
    # Add Metadata section
    metadata = etree.SubElement(root, "Metadata")
    creator = etree.SubElement(metadata, "Creator")
    creator.text = "Transkribus"
    created = etree.SubElement(metadata, "Created")
    created.text = datetime.now().isoformat()
    last_change = etree.SubElement(metadata, "LastChange")
    last_change.text = datetime.now().isoformat()
    
    # Add TranskribusMetadata (example values, adjust as needed)
    transkribus_metadata = etree.SubElement(metadata, "TranskribusMetadata", 
        docId="5672972", pageId="86044967", pageNr="1", tsid="211150608", 
        status="IN_PROGRESS", userId="93284", 
        imgUrl="https://files.transkribus.eu/Get?id=XDAZLXELJGNSBBTYYDXGGHZE&fileType=view", 
        xmlUrl="https://files.transkribus.eu/Get?id=FEXQJQSQNEQBZIZRROJMOSYG", 
        imageId="71265050"
    )
    
    # Add Page element
    page = etree.SubElement(root, "Page", 
        imageFilename=image_data['filename'], 
        imageWidth=str(image_data['size']), 
        imageHeight=str(image_data['size'])
    )
    
    # Create ReadingOrder element
    reading_order = etree.SubElement(page, "ReadingOrder")
    ordered_group = etree.SubElement(reading_order, "OrderedGroup", id=f"ro_{image_id}", caption="Regions reading order")
    
    # Iterate through the regions and add them to the XML
    for idx, region in enumerate(image_data['regions'], start=1):
        # Add region reference for reading order
        region_ref = etree.SubElement(ordered_group, "RegionRefIndexed", index=str(idx - 1), regionRef=f"region_{idx}")
        
        # Calculate points based on the bounding box
        x = region['shape_attributes']['x']
        y = region['shape_attributes']['y']
        width = region['shape_attributes']['width']
        height = region['shape_attributes']['height']
        
        points = f"{x},{y} {x + width},{y} {x + width},{y + height} {x},{y + height}"
        
        # Add TextRegion element
        text_region = etree.SubElement(page, "TextRegion", 
            type=class_mapping.get(region['region_attributes']['Classes'], "Unknown"), 
            id=f"region_{idx}", 
            custom=f"readingOrder {{index:{idx - 1};}} structure {{type:{class_mapping.get(region['region_attributes']['Classes'], 'Unknown')}}};"
        )
        
        # Add Coords element
        coords = etree.SubElement(text_region, "Coords", points=points)

    # Return the XML string with the declaration
    return xml_declaration + etree.tostring(root, pretty_print=True, encoding='unicode')

# Iterate through the image metadata and create XML files
output_dir = os.path.join(PATH_DATA, "xml")  # Specify your output directory
os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

for image_id, image_data in data['_via_img_metadata'].items():
    xml_content = create_xml(image_id, image_data)  # Use image data directly
    
    # Debugging: Check if xml_content is None
    if xml_content is None:
        print(f"Failed to create XML for image ID: {image_id}")
        continue  # Skip to the next image if XML creation failed
    
    xml_file_path = os.path.join(output_dir, f"{image_data['filename'].split('.')[0]}.xml")
    
    # Save the XML content with UTF-8 encoding
    with open(xml_file_path, 'w', encoding='utf-8') as xml_file:
        xml_file.write(xml_content)

print("XML files created successfully.")
```
