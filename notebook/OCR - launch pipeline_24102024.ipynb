## 4. Launch Parsing

Objectives: parse all the images in the PAGES folder, save the results in the XML, and lines folders. 

    - PAGES: folder/subfolders containing images 
    - XML: folder to contain the XML files and store in corresponding subfolders
    - lines: folder to contain the lines images and store in corresponding subfolders

Methodology of parsing: 

    - parsing is done line by line, columns are reconstructed based on the XML file
    - for each page, the XML file contains the coordinates of the columns and the lines
    - for each page, the parsed lines are stored as seperate images in LINES/subfolder/PAGE/column_X/line_Y.jpg
    - each line is saved in a separate image in a column subfolder in the LINES folder

WARNING: parsing is currently done by batch. could be done page by page

```python
import subprocess
import importlib.util
import sys


# List all subfolders in data/mar_hopitaux_deces
#subfolders = [f for f in os.listdir(PATH_DATA) if os.path.isdir(join(PATH_DATA, f))]
subfolders = ['test']
print("Subfolders in data/mar_hopitaux_deces:")
for folder in subfolders:
    print(f"- {folder}")

# Add PATH_UTILS to sys.path
PATH_PARSE = join(PATH_REPO, 'trocr_handwritten', 'parse')
sys.path.append(PATH_PARSE)

# Process each subfolder
for subfolder in subfolders:
    print(f"\nProcessing subfolder: {subfolder}")
    
    # Define paths for this subfolder
    PATH_PAGES = join(PATH_DATA, subfolder, 'PAGES')
    PATH_XML = join(PATH_DATA, subfolder, 'XML')
    PATH_LINES = join(PATH_DATA, subfolder, 'lines')
    
    # Create directories if they don't exist
    os.makedirs(PATH_XML, exist_ok=True)
    os.makedirs(PATH_LINES, exist_ok=True)

    # Define the command as a string
    command = f"python trocr_handwritten/parse/parse_page.py 
    --PATH_PAGES {PATH_PAGES} 
    --PATH_MODELS {PATH_MODELS} 
    --PATH_XML {PATH_XML} 
    --PATH_LINES {PATH_LINES} --verbose TRUE"

    print(f"Executing command for {subfolder}...")
    
    # Execute the command and capture the output
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)

    # Print the output line by line
    for line in process.stdout:
        print(line.decode('utf-8', errors='ignore').strip())

    # Wait for the process to finish and get the return code
    process.wait()
    return_code = process.returncode

    # Check if the command executed successfully
    if return_code == 0:
        print(f"Command executed successfully for {subfolder}")
    else:
        print(f"Command failed for {subfolder} with return code {return_code}")
```

### Consolidated

```python
for my_page in pages:
    _preds = pages[my_page]['preds']
    id_columns = list(pages[my_page]['columns'].keys())
    id_columns.sort()
    for column_id in id_columns:
        with open(join(PATH_PREDS, f'{my_page}_column_{column_id}.txt'), 'a') as f:
            for text in _preds.get(column_id):
                f.write(text)
                f.write('\n')
            f.close()
            
import pickle as pkl

pkl.dump(pages, open(join(PATH_PREDS, 'pages_predictions.pkl'), 'wb'))
```

### By lines

```python
import pandas as pd

for my_page in pages:

    try:
        image = Image.open(join(PATH_PAGES,f'{my_page}.jpg')).convert("RGB")
    except:
        image = Image.open(join(PATH_PAGES,f'{my_page}.JPG')).convert("RGB")

    image = image.convert("RGB")
    
    image_lines = pages[my_page]['lines']
    preds = pages[my_page]['lines_preds']
    image_coords = pages[my_page]['lines_coords']

    data_preds = pd.DataFrame([image_lines, preds, image_coords]).T
    data_preds.columns = ['file_name', 'text', 'coords']

    images = []
    for coord in data_preds.coords.values:
        images.append(image.crop(coord))

    images[0].save(join(PATH_LINES, f"{my_page}_lines.pdf"), save_all=True, append_images=images[1:])
    data_preds.to_csv(join(PATH_LINES, f"{my_page}_lines_preds.txt"), index=False, sep='\t')
```
